# **Removing Outliers**

```python
# Removing Outliers

data_crop = data_crop[data_crop['Area'] > 100.0]
data_crop = data_crop[data_crop['Production'] > 100.0]
data_crop.describe().apply(lambda s: s.apply(lambda x: format(x, 'f')))
```

# **Removing Rows with Null Values**

```python
#Removing Rows with Null Values
#.dropna(axis=0) to remove any rows with null values.

data_crop =data_crop.dropna(axis=0)
data_crop.shape
```

# **Data Reduction**

```python
#Removing Redundant variables
#First I got rid of variables that I thought were redundant. This include 'District','Area Units','Production Units'

data_crop = data_crop.drop(['District','Area Units','Production Units'], axis=1)
data_crop
```

# **Pairplot**

```python
#LabelEncoder can be used to normalize labels. It can also be used to transform non-numerical labels to numerical labels.

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
labelencoder= LabelEncoder()
```

```python
data_crop['State'] = labelencoder.fit_transform(data_crop['State'])
data_crop['Crop'] = labelencoder.fit_transform(data_crop['Crop'])
data_crop['Season'] = labelencoder.fit_transform(data_crop['Season'])
data_crop['Year'] = labelencoder.fit_transform(data_crop['Year'])
```

```python
sns.set()
columns = ['State','Crop','Season','Area','Production','Yield']
data_crop = pd.DataFrame(data_crop, columns=columns)
sns.pairplot(data_crop [columns], height = 3 ,kind ='scatter',diag_kind='kde')
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4),(ax5,ax6)) = plt.subplots(nrows=3, ncols=2, figsize=(10,10))
State_scatter_plot = pd.concat([data_crop['Crop'],data_crop['State'].astype('float')],axis = 1)
sns.regplot(x='State',y = 'Crop',data = State_scatter_plot,scatter= True, fit_reg=True, ax=ax1)
Season_scatter_plot = pd.concat([data_crop['Crop'],data_crop['Season']],axis = 1)
sns.regplot(x='Season',y = 'Crop',data = Season_scatter_plot,scatter= True, fit_reg=True, ax=ax2)
Area_scatter_plot = pd.concat([data_crop['Crop'],data_crop['Area']],axis = 1)
sns.regplot(x='Area',y = 'Crop',data = Area_scatter_plot,scatter= True, fit_reg=True, ax=ax3)
Production_scatter_plot = pd.concat([data_crop['Crop'],data_crop['Production']],axis = 1)
sns.regplot(x='Production',y ='Crop',data = Production_scatter_plot,scatter= True, fit_reg=True, ax=ax4)
Yield_scatter_plot = pd.concat([data_crop['Crop'],data_crop['Yield']],axis = 1)
sns.regplot(x='Yield',y = 'Crop',data = Yield_scatter_plot,scatter= True, fit_reg=True, ax=ax5)
```
